```python
#Data Preprocessing:
import pandas as pd

#load the datase/
data = pd.read_csv("C:/Users/LICT/Downloads/data.csv")
print(data.head())
```

```python
#dropping irrelevant columns
data_cleaned= data.drop(columns=['id','Unnamed: 32'], errors='ignore')

#convert diagnosis to binary (M =1, B =0)
data_cleaned['diagnosis'] = data_cleaned['diagnosis'].map({'M':1, 'B':0})

# Check the cleaned dataset
print(data_cleaned.head())
```

```python
x=data_cleaned.drop(columns='diagnosis')
y=data_cleaned['diagnosis']

#Display the shapes of x and y
print("Features shape:", x.shape)
print("Target shape:", y.shape)
```

```python
# Import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
# Import train_test_split
from sklearn.model_selection import train_test_split
# Import accuracy_score
from sklearn.metrics import accuracy_score
#Split the dataset into 80% train, 20% test
x_train, x_test, y_train, y_test = train_test_split(x,y,
                                                   test_size=0.2,
                                                   stratify=y,
                                                   random_state=1)
##checking the shape of the split
print("Training features shape:", x_train.shape)
print("Testing features shape:", x_test.shape)
```

```python
# Import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
# Import train_test_split
from sklearn.model_selection import train_test_split
# Import accuracy_score
from sklearn.metrics import accuracy_score
#Split the dataset into 80% train, 20% test
x_train, x_test, y_train, y_test = train_test_split(x,y,
                                                   test_size=0.2,
                                                   stratify=y,
                                                   random_state=1)
#Instantiate and train the decision tree model 
dt_model = DecisionTreeClassifier(max_depth=2, random_state=1)

#Fit dt_model to the training set
dt_model.fit(x_train, y_train)

#Predict the test set labels
y_pred = dt_model.predict(x_test)

# Evaluate the test-set accuracy
accuracy_score(y_test, y_pred)
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from matplotlib.colors import ListedColormap

# Generate a synthetic dataset with two features
X, y = make_classification(n_features=2, n_informative=2, n_redundant=0, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize classifiers
logreg = LogisticRegression(random_state=1)
dt = DecisionTreeClassifier(random_state=1)

# Train classifiers
logreg.fit(X_train, y_train)
dt.fit(X_train, y_train)

# Define the plotting function
def plot_labeled_decision_regions(X, y, classifiers):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))

    for clf in classifiers:
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        plt.contourf(xx, yy, Z, alpha=0.3, cmap=ListedColormap(['red', 'blue']))

    plt.scatter(X[:, 0], X[:, 1], c=y, s=40, edgecolor='k', cmap=ListedColormap(['red', 'blue']))
    plt.title("Decision Regions")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()

# Define the list of classifiers
clfs = [logreg, dt]

# Plot decision regions
plot_labeled_decision_regions(X_test, y_test, clfs)
```
