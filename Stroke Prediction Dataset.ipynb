# STATISTICAL TESTING

**Chi-square Test**

```python
from scipy.stats import chi2_contingency

# Example: Chi-Square test for gender and Churn
contingency_table = pd.crosstab(data['gender'], data['stroke'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
if p < 0.05:
    print("There is a significant association between gender and stroke.")
else:
    print("There is no significant association between gender and stroke.")
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

# Load your dataset (replace with your dataset)
# Example: data = pd.read_csv('your_dataset.csv')
# Ensure 'data' is defined and contains the relevant columns

# List of features and target variable
features = ['id', 'gender', 'age', 'hypertension', 'heart_disease', 'ever_married', 'Residence_type', 
            'avg_glucose_level', 'bmi', 'work_type_Never_worked', 'work_type_Private', 
            'work_type_Self-employed', 'work_type_children', 'smoking_status_formerly smoked', 
            'smoking_status_never smoked', 'smoking_status_smokes']  # Replace with your features
target = 'stroke'  # Replace with your target column

# Initialize a list to store results
results = []

for feature in features:
    try:
        # Create a contingency table
        contingency_table = pd.crosstab(data[feature], data[target])

        # Perform the Chi-Square test
        chi2, p, dof, expected = chi2_contingency(contingency_table)

        # Determine significance
        significant = p < 0.05  # Assuming 0.05 as the significance threshold

        # Append the results
        results.append({
            'Feature': feature,
            'Chi2_Statistic': chi2,
            'P_Value': p,
            'Significant': significant
        })

    except Exception as e:
        print(f"Error with feature '{feature}': {e}")

# Convert results to a DataFrame
results_df = pd.DataFrame(results)

# Print the results
print("Chi-Square Test Results:")
print(results_df)
```

**Anova**

```python
from scipy.stats import f_oneway
import pandas as pd

# Specify the numerical and categorical variables
numerical_features = ['age', 'avg_glucose_level', 'bmi']  # Replace with your numerical features
grouping_variable = 'stroke'  # Replace with your grouping variable

# Iterate over numerical features and perform ANOVA
for feature in numerical_features:
    try:
        # Group data by the categorical variable (e.g., stroke groups)
        grouped_data = data.groupby(grouping_variable)[feature]

        # Extract the values for each group
        groups = [grouped_data.get_group(value) for value in grouped_data.groups.keys()]

        # Perform ANOVA
        f_stat, p_value = f_oneway(*groups)

        # Print results
        print(f"ANOVA Results for {feature}:")
        print(f"F-statistic: {f_stat}")
        print(f"P-value: {p_value}")
        if p_value < 0.05:
            print(f"There is a significant difference in {feature} between stroke groups.\n")
        else:
            print(f"There is no significant difference in {feature} between stroke groups.\n")

    except Exception as e:
        print(f"Error analyzing feature '{feature}': {e}")
```
