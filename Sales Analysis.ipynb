```python
import os
import pandas as pd

# Define the folder path (use raw string or forward slashes)
folder_path = r"C:\Users\LICT\SalesAnalysis\Sales_Data"

# Get the list of all files in the Sales_Data directory
files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Initialize an empty DataFrame to store all the data
all_months_data = pd.DataFrame()

# Loop through each file and read it into the DataFrame
for file in files:
    # Correctly join folder path with the file name
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    all_months_data = pd.concat([all_months_data, df], ignore_index=True)

# Display the first few rows of the combined data
all_months_data.head()
```

# Question 4: What products are most often sold together? 

```python
all_data.head()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df.head(20)
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
for key,value in count.most_common(10):
    print(key, value)
```
