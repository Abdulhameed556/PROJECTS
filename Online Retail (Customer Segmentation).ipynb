# Feature Engineering

**Create Total Spending**

```python
data['TotalPrice'] = data['Quantity'] * data['UnitPrice']
total_spending = data.groupby('CustomerID')['TotalPrice'].sum()
```

```python
total_spending.head()
```

**Purchase Frequency**

```python
#  Count how many transactions each customer made:
purchase_freq = data.groupby('CustomerID')['InvoiceNo'].nunique()
purchase_freq.head()
```

**Recency**

```python
# Use the most recent date and calculate days since last purchase:
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
last_purchase = data.groupby('CustomerID')['InvoiceDate'].max()
recency = (data['InvoiceDate'].max()-last_purchase).dt.days

recency
```

***Combine Features**

```python
customer_data = pd.DataFrame({'TotalSpending':total_spending,
                             'PurchaseFrequency':purchase_freq,
                             'Recency':recency})
customer_data.reset_index(inplace=True)
```

# Apply Clustering Algorithms

Clustering is sensitive to scale, so standardize the features:

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_data[['TotalSpending', 'PurchaseFrequency', 'Recency']])
```

**K-Means Clustering**

Use the Elbow Method to determine the optimal number of clusters

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
    
plt.plot(range(1, 10), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')    
```

```python
# Fit KMeans with the optimal k
kmeans = KMeans(n_clusters=3, random_state=42)
customer_data['Segment'] = kmeans.fit_predict(scaled_data)
```

**Visualize Segments**



```python
import seaborn as sns
sns.scatterplot(x='TotalSpending', y='Recency', hue='Segment', data=customer_data, palette='viridis')
plt.title('Customer Segments')
plt.show()
```

Splitting data into time frames (e.g., monthly) for trend analysis.

```python
# Convert InvoiceDate to datetime 
data['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Create a new column for month-year
data['MonthYear'] = data['InvoiceDate'].dt.to_period('M')

# Group data by customer and month-year
monthly_summary = data.groupby(['CustomerID', 'MonthYear']).agg({'InvoiceNo': 'nunique',
                                                               'Quantity': 'sum',    # Number of orders
                                                               'TotalPrice': 'sum'   # Total quantity purchased
                                                               }).reset_index()      # Total spending (calculated earlier)
# View the monthly summary
print(monthly_summary.head())
```

# Classifying Customers


**Splitting Data**


```python
from sklearn.model_selection import train_test_split

# Features (spending, number of orders, etc.)
X = customer_summary[['TotalPrice', 'NumOrders', 'AvgBasketSize']]

# Target (customer cluster)
y = customer_summary['Cluster']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"Training set size: {X_train.shape[0]}, Testing set size: {X_test.shape[0]}")
```

**Support Vector Classifier (SVC)**

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

# Train the SVC model
svc_model = SVC(kernel='linear', random_state=42)
svc_model.fit(X_train, y_train)

# Predict on the test set
y_pred_svc = svc_model.predict(X_test)

# Evaluate the model
print("SVC Model Evaluation")
print(confusion_matrix(y_test, y_pred_svc))
print(classification_report(y_test, y_pred_svc))
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

# Train the Logistic Regression model
log_reg = LogisticRegression(random_state=42)
log_reg.fit(X_train, y_train)

# Predict on the test set
y_pred_log = log_reg.predict(X_test)

# Evaluate the model
print("Logistic Regression Evaluation")
print(confusion_matrix(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier

# Train the Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Predict on the test set
y_pred_dt = dt_model.predict(X_test)

# Evaluate the model
print("Decision Tree Evaluation")
print(confusion_matrix(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))
```

**Voting Classifier**

Combine multiple models for better predictions using a voting ensemble.

```python
from sklearn.ensemble import VotingClassifier

# Combine models into a voting ensemble
voting_model = VotingClassifier(estimators=[('svc', svc_model),
                               ('log_reg', log_reg),
                               ('rf', rf_model)],
                                voting='hard')

# Train the ensemble model
voting_model.fit(X_train, y_train)

#Predict on the test set
y_pred_voting = voting_model.predict(X_test)

# Evaluate the model
print("Voting Classifier Evaluation")
print(confusion_matrix(y_test, y_pred_voting))
print(classification_report(y_test, y_pred_voting))


                                
                            
```

```python
from sklearn.metrics import classification_report, roc_auc_score, accuracy_score

# Initialize a dictionary to store results
evaluation_results = {}

# Function to evaluate a model
def evaluate_model(name, model, X_test, y_test):
    y_pred = model.predict(X_test)
    
    # Accuracy, Precision, Recall, F1-score
    report = classification_report(y_test, y_pred, output_dict=True)
    accuracy = accuracy_score(y_test, y_pred)
    
    # AUC Score
    try:
        if len(set(y_test)) > 2:  # Multi-class case
            y_prob = model.predict_proba(X_test)
            auc = roc_auc_score(y_test, y_prob, multi_class="ovr")  # Use 'ovr' strategy
        else:  # Binary classification case
            y_prob = model.predict_proba(X_test)[:, 1]
            auc = roc_auc_score(y_test, y_prob)
    except AttributeError:
        auc = "N/A"  # Some models like SVC may not have predict_proba
    
    # Store the results
    evaluation_results[name] = {
        "Accuracy": accuracy,
        "Precision": report['weighted avg']['precision'],
        "Recall": report['weighted avg']['recall'],
        "F1-Score": report['weighted avg']['f1-score'],
        "AUC": auc
    }

# Evaluate all models
evaluate_model("SVC", svc_model, X_test, y_test)
evaluate_model("Logistic Regression", log_reg, X_test, y_test)
evaluate_model("k-NN", Knn_model, X_test, y_test)
evaluate_model("Decision Tree", dt_model, X_test, y_test)
evaluate_model("Random Forest", rf_model, X_test, y_test)
evaluate_model("AdaBoost", ab_model, X_test, y_test)
evaluate_model("Gradient Boosting", gb_model, X_test, y_test)
evaluate_model("Voting Classifier", voting_model, X_test, y_test)
```

```python
import pandas as pd

# Create a DataFrame from evaluation results
results_df = pd.DataFrame(evaluation_results).T

# Sort models by Accuracy
results_df = results_df.sort_values(by="Accuracy", ascending=False)

# Display the table
print("Model Evaluation Results:")
print(results_df)
```
